```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import os
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df.loc[df.h_shot_acc > 100].head()
```

```python
full_match_df = df.loc[(df.time == "90:00") | (df.time == "120:00")]
len(full_match_df)
```

```python
full_match_df.head()
```

```python
home_winner = full_match_df.loc[full_match_df.h_score > full_match_df.a_score]
more_shots_count = len(home_winner.loc[home_winner.h_shots > home_winner.a_shots])
more_shots_on_target_count = len(home_winner.loc[home_winner.h_shots_on_target > home_winner.a_shots_on_target])
more_possession_count = len(home_winner.loc[home_winner.h_possession > home_winner.a_possession])
more_tackles_count = len(home_winner.loc[home_winner.h_tackles > home_winner.a_tackles])
more_fouls_count = len(home_winner.loc[home_winner.h_fouls > home_winner.a_fouls])
more_corners_count = len(home_winner.loc[home_winner.h_corners > home_winner.a_corners])

away_winner = full_match_df.loc[full_match_df.h_score < full_match_df.a_score]
more_shots_count = more_shots_count + len(away_winner.loc[away_winner.h_shots < away_winner.a_shots])
more_shots_on_target_count = more_shots_on_target_count + len(away_winner.loc[away_winner.h_shots_on_target < away_winner.a_shots_on_target])
more_possession_count = more_possession_count + len(away_winner.loc[away_winner.h_possession < away_winner.a_possession])
more_tackles_count = more_tackles_count + len(away_winner.loc[away_winner.h_tackles < away_winner.a_tackles])
more_fouls_count = more_fouls_count + len(away_winner.loc[away_winner.h_fouls < away_winner.a_fouls])
more_corners_count = more_corners_count + len(away_winner.loc[away_winner.h_corners < away_winner.a_corners])

total_winners = len(home_winner) + len(away_winner)
print("Out of " + str(total_winners) + " winning teams:\n - " + 
      "{:.2f}% had more shots\n - ".format(100*(more_shots_count/total_winners)) +
      "{:.2f}% had more shots on target\n - ".format(100*(more_shots_on_target_count/total_winners)) +
      "{:.2f}% had more possession\n - ".format(100*(more_possession_count/total_winners)) +
      "{:.2f}% had more tackles\n - ".format(100*(more_tackles_count/total_winners)) +
      "{:.2f}% had more fouls\n - ".format(100*(more_fouls_count/total_winners)) +
      "{:.2f}% had more corners".format(100*(more_corners_count/total_winners)))
```

```python
from scipy.stats import pearsonr
```

```python
clear_home_shots_df = full_match_df.loc[(full_match_df.h_shots >= 0) & (full_match_df.h_score >= 0)]
clear_away_shots_df = full_match_df.loc[(full_match_df.a_shots >= 0) & (full_match_df.a_score >= 0)]

shots_list = np.append(clear_home_shots_df.h_shots.to_numpy(), clear_away_shots_df.a_shots.to_numpy())
score_list = np.append(clear_home_shots_df.h_score.to_numpy(), clear_away_shots_df.a_score.to_numpy())
```

```python
pearsonr(shots_list, score_list)
```
