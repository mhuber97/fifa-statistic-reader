# Keras OCR

```python
import keras_ocr
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# keras-ocr will automatically download pretrained
# weights for the detector and recognizer.
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
image_path = "data/10-11-2020_16-35-35-qyrqt4lh.png"

image = cv2.imread(image_path)
```

```python
prediction_groups = pipeline.recognize([image])
```

```python
detections = prediction_groups
```

```python
for i in detections[0]:
    detected_coordinates = np.array(i[1], dtype="int")
    start_coordinate = (detected_coordinates[0][0], detected_coordinates[0][1])
    end_coordinate = (detected_coordinates[2][0], detected_coordinates[2][1])
    image = cv2.rectangle(image, start_coordinate, end_coordinate, (255, 0, 0), 2)
```

```python
plt.rcParams['figure.dpi'] = 150
plt.imshow(image)
```
